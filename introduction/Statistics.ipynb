```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

-**Numpy(Numerical Python)** is the library
use for Efficient operations on arrays.

-**Pandas** handles dataset in tabular form
-**Matplotlib and Seaborn** is use for creating visualization

-**Scipy(Scientific Python)** performs Statistical Computation




 **Descriptive Statistics**

```python
data =[10,20,15,25,30,35,40,45,40,30]

mean = np.mean(data)
print(f"Mean: {mean}")

median = np.median(data)
print(f"Median:{median}")

mode = stats.mode(data)
print(f"Mode: {mode}")


std_dev= np.std(data)
print(f"Standard Deviation: {std_dev}")

variance = np.var(data)
print(f"Variance:{variance}")
```

**Data Visualization**

```python
plt.figure(figsize=(8 ,4))

plt.hist(data, bins=5, color ="red", alpha =0.7,edgecolor ="black")
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

sns.boxplot(data=data, color ="lightblue")
plt.title("Box plot")
plt.show()
```

**Probability Distribution**

```python
normal_data = np.random.normal(loc=50, scale=10, size=1000)

sns.histplot(normal_data, kde=True, color="green")
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

**Inferential Statistics**

```python
confidence = 0.95
n = len(data)
mean = np.mean(data)
std_err= stats.sem(data)
ci = stats.t.interval(confidence, df=n-1, loc=mean,scale= std_err)

print(f"95% of confidence Interval:{ci}")
```

**Hypothesis Testing**

```python
sample_mean = np.mean(data)
population_mean = 30
t_stat, p_value = stats.ttest_1samp(data, population_mean)

print(f"T-statistics:{t_stat}, P-value: {p_value}")
if p_value < 0.05:
  print("Reject the null hypothesis")
else:
  print("Fail to reject the null hypothesis")
```

**Correlation**

```python
data = pd.DataFrame({
    "X": [1, 2, 3, 4, 5],
    "Y": [2, 4, 5, 4, 5]
})

# Correlation Coefficient
correlation = data["X"].corr(data["Y"])
print(f"Correlation Coefficient: {correlation}")
```

**Linear Regression**

```python
from sklearn.linear_model import LinearRegression

# Features and target
X = np.array(data["X"]).reshape(-1, 1)  # Reshaping for scikit-learn
y = np.array(data["Y"])

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Regression Coefficients
print(f"Slope: {model.coef_[0]}, Intercept: {model.intercept_}")

# Predictions
y_pred = model.predict(X)
print(f"Predictions: {y_pred}")
```
